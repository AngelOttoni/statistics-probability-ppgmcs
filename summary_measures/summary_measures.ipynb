>Os cálculos de média e desvio-padrão realizados no Python foram baseados nos dados amostrais fornecidos (n=50 vacas). Portanto, os resultados representam estimativas da média e da dispersão da população total de vacas da raça Holandesa. Para o desvio-padrão, foi utilizada a fórmula com correção para amostras (graus de liberdade n - 1), que é o procedimento padrão nos métodos estatísticos para dados amostrais.

```python
# Lista dos valores fornecidos
dados = [
    2825, 2072, 2733, 2069, 2484, 4285, 2862, 3353, 1449, 2029, 1258, 2982,
    2045, 1677, 1619, 2597, 3512, 2444, 1773, 2284, 1884, 2359, 2046, 2364,
    2669, 3109, 2804, 1658, 2207, 2159, 2207, 2882, 1647, 2051, 2202, 3223,
    2383, 1732, 2230, 1147, 2711, 1874, 1979, 1319, 2923, 2281, 1230, 1665,
    1294, 2936
]

# Converte para Series
prod_mensal_leite = pd.Series(dados)
```

## **2.1. Encontre a média *amostral* dos dados.**

```python
media = prod_mensal_leite.mean()

print(f'Média amostral: {media:.2f} libras')
```

## **2.2. Encontre o desvio-padrão *amostral* dos dados.**

```python
desvio_padrao = prod_mensal_leite.std()  # padrão amostral

print(f'Desvio-padrão amostral: {desvio_padrao:.2f} libras')
```

## **2.3. Faça a distribuição de frequência para os dados, utilizando largura 500 para as classes.**

```python
# Defini os intervalos de classe
classes = range(1000, 5000, 500)

# Frequência por classe
frequencia = pd.cut(prod_mensal_leite, bins=classes).value_counts().sort_index()

# Tabela de frequência
tabela_freq = pd.DataFrame({
    'Intervalo': frequencia.index.astype(str),
    'Frequência': frequencia.values
})
tabela_freq
```

## **2.4. Faça um histograma para os dados. A distribuição parece ter forma de sino?**

```python
plt.figure(figsize=(10, 6))
sns.histplot(prod_mensal_leite, bins=classes, kde=False, color='skyblue', edgecolor='black')
plt.title('Histograma da produção de leite')
plt.xlabel('Produção mensal (libras)')
plt.ylabel('Frequência')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

>Observando o histograma, é possível perceber que a distribuição tem um formato aproximadamente semelhante ao de uma curva normal (forma de sino), com a maioria dos dados concentrados no intervalo central (entre 2000 e 3000 libras). No entanto, nota-se uma leve assimetria à direita, com alguns valores mais afastados na cauda superior (acima de 3000 libras). Portanto, podemos dizer que a distribuição não é perfeitamente simétrica, mas se aproxima de uma forma de sino, característica de distribuições aproximadamente normais.

## **2.5. Qual porcentagem da distribuição está dentro de um desvio-padrão em relação à média? Dentro de dois desvios-padrão em relação à média? Como esses resultados concordam com a regra empírica?**

```python
# Calcula os limites
lim_1 = (media - desvio_padrao, media + desvio_padrao)
lim_2 = (media - 2*desvio_padrao, media + 2*desvio_padrao)

# Porcentagens reais nos intervalos
dentro_1dp = prod_mensal_leite.between(*lim_1).mean() * 100
dentro_2dp = prod_mensal_leite.between(*lim_2).mean() * 100

print(f'Porcentagem dentro de 1 DP: {dentro_1dp:.2f}%')
print(f'Porcentagem dentro de 2 DP: {dentro_2dp:.2f}%')
```

### **2.6. Utilize a distribuição de frequência para estimar a média amostral dos dados. Compare seus resultados com o Exercício 1.**

```python
# Calcula o ponto médio (xi) de cada intervalo
intervalos = frequencia.index
ponto_medio = [(interval.left + interval.right) / 2 for interval in intervalos]
fi = frequencia.values

# Tabela com ponto médio (xi) e frequência (fi)
tabela_agrupada = pd.DataFrame({
    'Classe': [str(i) for i in intervalos],
    'xi': ponto_medio,
    'fi': fi
})

tabela_agrupada
```

```python
# Média estimada
numerador_media = (tabela_agrupada['fi'] * tabela_agrupada['xi']).sum()
denominador = tabela_agrupada['fi'].sum()
media_agr = numerador_media / denominador

print(f'Média estimada dos dados agrupados: {media_agr:.2f} libras')
```

>Utilizando a distribuição de frequência dos dados, estimei a média amostral por meio da média ponderada dos pontos médios de cada classe. Para isso, associei a cada intervalo de classe o seu ponto médio (xi) e multipliquei pela respectiva frequência (fi). Em seguida, somei os produtos e dividi pelo total de observações. O cálculo resultou em uma média estimada de 2270,00 libras. Comparando com a média amostral real calculada no Exercício 1, que foi de 2270,54 libras, observa-se que os valores são similares. Essa pequena diferença de 0,54 libra é esperada, pois a média estimada usa os pontos médios como aproximação para os valores reais dentro de cada classe. Ainda assim, o resultado demonstra que a distribuição de frequência representa bem os dados originais. Portanto, a média estimada a partir da tabela agrupada confirma a média amostral obtida com os dados brutos, validando a consistência da análise.


### **2.7. Utilize a distribuição de frequência para encontrar o desvio-padrão amostral para os dados. Compare seus resultados com o Exercício 2.**

```python
# Variância estimada
tabela_agrupada['(xi - média)^2'] = (tabela_agrupada['xi'] - media_agr) ** 2
tabela_agrupada['fi*(xi - média)^2'] = tabela_agrupada['fi'] * tabela_agrupada['(xi - média)^2']

variancia_agr = tabela_agrupada['fi*(xi - média)^2'].sum() / (denominador - 1)
desvio_agr = np.sqrt(variancia_agr)

print(f'Desvio-padrão estimado: {desvio_agr:.2f} libras')
```

>Utilizando a distribuição de frequência dos dados, estimei o desvio-padrão amostral aplicando a fórmula do desvio-padrão para dados agrupados. Para isso, considerei os pontos médios de cada classe (xi), suas respectivas frequências (fi) e calculei a variância ponderada. Em seguida, apliquei a correção para amostras (divisão por $n - 1$) e extraí a raiz quadrada. O resultado foi um desvio-padrão estimado de 646,50 libras. Comparando com o valor obtido no Exercício 2, que foi de 653,18 libras (calculado diretamente com os dados individuais), observa-se que os dois resultados são muito próximos, com uma diferença de aproximadamente 6,68 libras. Essa pequena variação é esperada, já que o cálculo com dados agrupados utiliza os pontos médios como aproximação para todos os valores dentro de cada classe. Ainda assim, o resultado confirma que a distribuição de frequência oferece uma boa estimativa do desvio-padrão amostral, representando bem a dispersão dos dados.

### **b) Calcule os $\bar{x_t}$ utilizando $\alpha = 0,1$. Repita utilizando $\alpha = 0,5$. Que valor de $\alpha$ fornece uma série $\bar{x_t}$ mais suavizada?**

```python
# Suavização com alpha = 0.1
df_efluente['EWMA_0.1'] = df_efluente['Temperatura'].ewm(alpha=0.1, adjust=False).mean()

# Suavização com alpha = 0.5
df_efluente['EWMA_0.5'] = df_efluente['Temperatura'].ewm(alpha=0.5, adjust=False).mean()
```

```python
df_efluente.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_efluente.index + 1, df_efluente['Temperatura'], marker='o', label='Série Original', color='purple')
plt.plot(df_efluente.index + 1, df_efluente['EWMA_0.1'], marker='o', linestyle='--', label='EWMA (α=0.1)', color='green')
plt.plot(df_efluente.index + 1, df_efluente['EWMA_0.5'], marker='o', linestyle='--', label='EWMA (α=0.5)', color='orange')

plt.title('Suavização Exponencial da Temperatura do Efluente')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°F)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

> **A série com $\alpha = 0,1$ apresentou uma curva mais suave, com menos oscilações e maior resistência a variações bruscas de um dia para o outro. Isso ocorre porque um $\alpha$ menor dá mais peso ao histórico passado e menos peso ao valor mais recente. Por outro lado, a série com $\alpha = 0,5$ ficou mais próxima da série original, reagindo mais rapidamente a cada novo valor de temperatura.**

## **c) Substitua $\bar{x}_{t-1}= \alpha x_{t-1} + (1 - \alpha) x_{t-2}$ no lado direito da expressão de $\bar{x}_{t}$ e, então, substitua $\bar{x}_{t−2}$ em termos de ${x}_{t−2}$ e $\bar{x}_{t−3}$ e assim por diante. De quantos valores $x_t$, ${x}_{t−1}$, ... $x_1$ depende $\bar{x}_t$? O que acontece ao coeficiente de ${x}_{t−k}$ quando $k$ aumenta?**

```python
# Defini o alpha
alpha = 0.1
n = 14  # Número de observações (tamanho da série)

# Calcula os coeficientes teóricos para cada x_{t-k}
pesos = [alpha * (1 - alpha)**k for k in range(n)]

# Inverte a ordem para representar os pesos do x1 até xt
pesos_reverso = pesos[::-1]
```

```python
# Plotando os pesos
plt.figure(figsize=(8, 5))
plt.bar(range(1, n+1), pesos_reverso, color='purple', edgecolor='black')
plt.xlabel('Ordem da observação (k=1 é a mais antiga, k=14 é a mais recente)')
plt.ylabel('Peso na EWMA')
plt.title(f'Contribuição de cada x_k na EWMA com α={alpha}')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Exibi os pesos numéricos
for i, peso in enumerate(pesos_reverso, start=1):
    print(f'x_{i}: peso = {peso:.5f}')
```
